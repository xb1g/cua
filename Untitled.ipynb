```python
import json
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import Image, display
```

```python
with open("trajectories/aegis-rl-policy.json", "r") as f:
    policy = json.load(f)

policy
```

```python
stats = policy["stats"]

rows = []

for strategy, vals in stats.items():
    row = {"strategy": strategy}
    row.update(vals)
    rows.append(row)

df = pd.DataFrame(rows)

df["avg_reward"] = df["reward_sum"] / df["pulls"]

df
```

```python
plt.figure(figsize=(10,6))

plt.bar(df["strategy"], df["avg_reward"])

plt.title("Average Reward by Strategy")
plt.xlabel("Strategy")
plt.ylabel("Average Reward")

plt.xticks(rotation=15)

plt.show()
```

```python
plt.figure(figsize=(10,6))

plt.bar(df["strategy"], df["pulls"])

plt.title("Strategy Selection Frequency")
plt.xlabel("Strategy")
plt.ylabel("Pull Count")

plt.xticks(rotation=15)

plt.show()
```
